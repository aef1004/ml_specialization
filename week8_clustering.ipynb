## Clustering 
- finds similar datapoints
- ex. market segmentation, grouping similar views, DNA analysis


Step 1 - take a random guess at finding centers of cluster
Step 2 - assign each point to its nearest centroid (calculate squared difference between point and each centroid and choose
Step 3 - find average of each cluster and move centroid to that average 
Step 4 - interate through steps 2-4

Gaussian (normal) distributions (aka bell shaped curve)
 - centered at mu (avg) with standard deviation sigma

$\mu$ = (1/m) * sum(x(i)) aka average
variance = $\sigma^2$ = (1/m) * sum ( (xi - mu)^2  )



Gaussian distribution is given by

   $$ p(x ; \mu,\sigma ^2) = \frac{1}{\sqrt{2 \pi \sigma ^2}}\exp^{ - \frac{(x - \mu)^2}{2 \sigma ^2} }$$

   where $\mu$ is the mean and $\sigma^2$ is the variance.

## Anomaly detection vs supervised learning

### Anomaly detection
- very small # of positive examples (0-20 iscommon)
- large # of negative examples
- many different types of anomalies (future anomalies may look nothing like anomalous examples we've seen thus far)

Examples
- fraud detection
- manufacturing - finding unseen defects
- monitoring machines in data center

### Supervised 
Examples
- email spam classification
- manufacturing with previously seen defect
- weather prediction
- disease classification
